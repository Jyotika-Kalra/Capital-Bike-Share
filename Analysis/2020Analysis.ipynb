```python
df20["Start_Date"] = pd.to_datetime(df20["Start_Date"])
df20["End_Date"] = pd.to_datetime(df20["End_Date"])
```

```python
df20.dtypes
```

```python
# creating new column with day of the week
df20["Day_of_week"] = df20['Start_Date'].dt.strftime('%A')
```

```python
# Weekwise Bike count
day20 = df20["Day_of_week"].value_counts()
day20
```

```python
#Sorting week days in order
weekday= ["Monday","Tuesday","Wednesday", "Thursday","Friday","Saturday","Sunday"]
day20.index = pd.CategoricalIndex(day20.index, categories=weekday, ordered=True)
day20 = day20.sort_index()
```

```python
#Plotting in Bar Graph
plt.figure(figsize=(10,8))
day20.plot(kind="bar", edgecolor='Black')
plt.style.use("seaborn-pastel")
plt.title("Rides per weekday")
plt.ylabel('Count')
plt.tight_layout()
```
